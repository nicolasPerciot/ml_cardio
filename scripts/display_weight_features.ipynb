```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

path_dir = os.path.dirname(os.path.abspath(''))

data = pd.read_csv(os.path.join(path_dir, 'data\\data_final_lite.csv')).drop(['ID', 'Congestive heart failure'], axis=1)
```

```python
data_first = data.iloc[:, :(round(len(data.columns.to_list()) / 2))].drop(['CHA2DS2-VASc'], axis=1)
data_first['CHA2DS2-VASc'] = data['CHA2DS2-VASc']
```

```python
data_sec = data.iloc[:, (round(len(data.columns.to_list()) / 2)):]
data_sec['CHA2DS2-VASc'] = data['CHA2DS2-VASc']
```

```python
import numpy as np

corr = data_first.corr()

mask = np.triu(np.ones_like(corr, dtype=bool))

f, ax = plt.subplots(figsize=(12,10))
cmap = sns.diverging_palette(230, 20, as_cmap=True)

sns.heatmap(corr, mask=mask, cmap=cmap,  center=0, square=True, linewidths=.5,cbar_kws={"shrink": .5},  annot=True)
```

```python
import numpy as np

corr = data_sec.corr()

mask = np.triu(np.ones_like(corr, dtype=bool))

f, ax = plt.subplots(figsize=(12,10))
cmap = sns.diverging_palette(230, 20, as_cmap=True)

sns.heatmap(corr, mask=mask, cmap=cmap,  center=0, square=True, linewidths=.5,cbar_kws={"shrink": .5},  annot=True)
```
